Create a function:
- if divisible by 3 print "Fizz"
- if divisible by 5 print "Buzz"
- if divisible by 3 and 5 print "Fizzbuzz"

```python
def fizzbuzz(n):
    for i in range(1, n+1, 1):
        result = ''
        if i % 3 == 0:
            result += 'Fizz'
        if i % 5 == 0:
            result += 'Buzz'
        print(i, result)
```

```python
fizzbuzz(100)
```
