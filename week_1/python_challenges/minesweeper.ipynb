```python
import pandas as pd
```

```python
mine_field = [[".",".",".",".",".","*",".","."], 
              [".","*",".",".",".",".",".","."],
              [".",".",".",".",".",".","*","."],
              [".",".","*","*",".",".",".","."],
              [".",".","*",".",".",".","*","."],
              [".",".",".",".",".",".",".","*"]]
```

```python
# 11101*10
# 1*100111
# 023211*1
# 02**1222
# 02*311*2
# 0111012*
```

```python
# correct wrong numbers and save as list

output = [["1","1","1","0","1","*","1","0"], 
          ["1","*","1","0","1","2","2","1"],
          ["1","2","3","2","1","1","*","1"],
          ["0","2","*","*","1","2","2","2"],
          ["0","2","*","3","1","1","*","2"],
          ["0","1","1","1","0","1","2","*"]]
```

```python
df_out = pd.DataFrame(output)
df_out
```

```python
WIDTH = len(mine_field[0])
HEIGHT = len(mine_field)
```

```python
df = pd.DataFrame(mine_field)
df
```

```python
df_zeroes = df.replace(to_replace=".", value=0)
df_zeroes
```

```python
for col_index in df_zeroes:
    counter = 0
    for tile in df_zeroes[col_index]:
        if tile == "*":
            # change tile directly above mine
            if counter > 0:
                if df_zeroes[col_index][counter - 1] != "*":
                    df_zeroes[col_index][counter - 1] += 1
            # change tile directly under mine
            if counter < HEIGHT - 1:
                if df_zeroes[col_index][counter + 1] != "*":
                    df_zeroes[col_index][counter + 1] += 1
            # change tile to the left
            if col_index > 0:
                if df_zeroes[col_index - 1][counter] != "*":
                    df_zeroes[col_index - 1][counter] += 1
            # change tile to the right
            if col_index < WIDTH - 1:
                if df_zeroes[col_index + 1][counter] != "*":
                    df_zeroes[col_index + 1][counter] += 1
            # change tile above and to the left
            if counter > 0 and col_index > 0:
                if df_zeroes[col_index - 1][counter - 1] != "*":
                    df_zeroes[col_index - 1][counter - 1] += 1
            # change tile below and to the left
            if counter > 0 and col_index < WIDTH - 1:
                if df_zeroes[col_index + 1][counter - 1] != "*":
                    df_zeroes[col_index + 1][counter - 1] += 1
            # change tile above and to the right
            if counter < HEIGHT - 1 and col_index > 0:
                if df_zeroes[col_index - 1][counter + 1] != "*":
                    df_zeroes[col_index - 1][counter + 1] += 1
            # change tile below and to the right
            if counter < HEIGHT - 1 and col_index < WIDTH - 1:
                if df_zeroes[col_index + 1][counter + 1] != "*":
                    df_zeroes[col_index + 1][counter + 1] += 1
        counter += 1
            
df_zeroes
```
