# Data Round-Trip

```python
import pandas as pd
```

## CSV

Read the CSV file `large_countries_2015.csv`

```python
df = pd.read_csv('large_countries_2015.csv', index_col=0)
df
```

## Excel

* Write the data into an Excel file
* Read the data again

```python
df.to_excel('output.xlsx')
df1 = pd.read_excel('output.xlsx', index_col=0)
df1
```

## JSON - Java Script Object Notation

* Write the data into a JSON file
* Read the data again

```python
df1.to_json('output.json')
df2 = pd.read_json('output.json')
df2
```
